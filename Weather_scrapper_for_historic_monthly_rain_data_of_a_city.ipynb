## **Weather scrapper**
### scraps from https://www.timeanddate.com/

### THIS SCRAPPER EXTRACTS HISTORIC RAIN DATA OF MORNING AND EVENING OF A CITY (MONTHLY BASIS)

SCRAPPER TAKES THE LINK OF TIMEANDDATE SITE OF THE INTERSTED CITY,INTERESTED MONTH AND YEAR.

IT THEN PRODUCES AND DOWNLOADS A CSV FILE OF THE DATA

IMPORTING LIBRARIES

```python
import json
from urllib.parse import urlencode

from bs4 import BeautifulSoup
import requests

import pandas as pd

import re
```

### GETTING WEBPAGE

```python
response = requests.get('https://www.timeanddate.com/weather/india/dehradun/historic?month=7&year=2022')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

Extracting tag that contain relevant info

```python
#<script type="text/javascript">
```

```python
test = soup.find_all('script',{'type':'text/javascript'})
```

```python
test
```

```python
type(test)
```

Finding relevant info

```python
testList = list(test[1])
testString = str(testList[0])
```

Creating a dictionary from relavant info

```python
dict_object = json.loads(re.search('({.+})', testString).group(0).replace("'", '"'))
```

```python
dict_object
```

```python
testList2 = list(dict_object)
```

```python
testList2
```

```python
print(dict_object['detail'][0])
```

```python
finalDict = {}

for i in range(1, len(dict_object['detail']),4):
  finalDict[dict_object['detail'][i]['ds']] = dict_object['detail'][i]['desc'].split(".")[0]
  finalDict[dict_object['detail'][i+1]['ds']] = dict_object['detail'][i+1]['desc'].split(".")[0]




```

Converting dictionary to list of list for conversion into csv file

```python
testList1 = []
for i in range(len(list(finalDict.keys()))):
  tempList = []
  
  for x in list(finalDict.keys())[i].split(','):
    tempList.append(x)

  tempList.append(finalDict[list(finalDict.keys())[i]])
  testList1.append(tempList)

testList1[0]
```

```python
testList1
```

## Exporting extracted data to csv file

```python
my_df = pd.DataFrame(testList1)
my_df.to_csv('Dehradun_Weather_july.csv', index=False, header=False)
```
